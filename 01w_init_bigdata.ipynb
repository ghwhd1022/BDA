### (4) 인구 데이터 전처리


```python
import pandas as pd
pop = pd.read_csv("/content/sample_data/popmonth.csv", encoding='cp949', index_col=0, thousands=',')
pop
```

### (5) 데이터프레임 pop의 열 이름 수정

```python
pop.columns = ['총인구수', '세대수', '세대당인구', '남자인구수', '여자인구수', '남여비율']
pop.head()
```

### (6) pop의 index를 수정

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

### (7) 인구 데이터 시각화 - '총인구수'의 산점도와 막대 그래프



```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(pop['총인구수'])
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pop1 = pop.sort_values('총인구수')
sns.barplot(x=pop1.index, y='총인구수', data=pop1)
plt.xticks(rotation=90)
plt.show()
```
